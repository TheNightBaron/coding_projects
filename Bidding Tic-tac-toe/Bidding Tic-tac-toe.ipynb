```python
class tic_tac_toe:
    def __init__(self):
        print('Enter mode')
        print('1. Normal')
        print('2. Betting')
        valid = False
        while not valid:
            self.mode = int(input())
            if ((self.mode > 2) or (self.mode < 1)):
                print('Invalid')
            else:
                valid = True
        self.board = ['-','-','-',
                '-','-','-',
                '-','-','-']
        self.game_still_going = True

        self.winner = None

        self.current_player = 'x'

        self.turn = 1
        
        self.player1 = self.player(str(input('Enter player 1 name ')),'x')
        print('{player} plays x'.format(player = self.player1.player_name))
        self.player2 = self.player(str(input('Enter player 2 name ')),'o')
        print('{player} plays o'.format(player = self.player2.player_name))

            
        
    def check_bet(self):
        if self.player1.bet > self.player2.bet:
            self.player1.tokens-=self.player1.bet
            self.player2.tokens+=self.player1.bet
            self.current_player='x'
            return 
        elif self.player1.bet < self.player2.bet:
            self.player2.tokens-=self.player2.bet
            self.player1.tokens+=self.player2.bet
            self.current_player='o'
            return
        else:
            print('Matching bets. Invalid')
            return

    def display_board(self):
        print(self.board[0] + '|' + self.board[1] + '|' + self.board[2])
        print(self.board[3] + '|' + self.board[4] + '|' + self.board[5])
        print(self.board[6] + '|' + self.board[7] + '|' + self.board[8])

    def flip_player(self):
        if self.current_player == 'x':
            self.current_player = 'o'
        else:
            self.current_player = 'x'
        return

    def handle_turn(self):
        if self.mode == 1:

            valid = False
            while not valid:
                position = int(input('Choose a position from 1-9:'))
                if ((position > 9) or (position < 1)):
                    print('Please input valid position')
                else:
                    position-=1
                    if self.board[position]!='-':
                        print('Illegal move; Try again!')
                    else:
                        self.board[position] = self.current_player
                        self.display_board()
                        valid = True
            return
        
        else:
            
            self.current_player = False
            while not self.current_player:
                valid = False
                while not valid:    
                    print('Player 1 balance = {balance}'.format(balance = self.player1.tokens))
                    print('Player 1 bet ')
                    self.player1.bet = int(input())
                    if ((self.player1.bet > self.player1.tokens) or (self.player1.bet < 0)):
                        print('Invalid')
                    else:
                        valid = True

                valid = False
                while not valid:
                    print('Player 2 balance = {balance}'.format(balance = self.player2.tokens))
                    print('Player 2 bet ')
                    self.player2.bet = int(input())
                    if ((self.player2.bet > self.player2.tokens) or (self.player2.bet < 0)):
                        print('Invalid')
                    else:
                        valid = True
                
                self.check_bet()
            
            valid = False
            while not valid:
                position = int(input('Choose a position from 1-9:'))
                if ((position > 9) or (position < 1)):
                    print('Please input valid position')
                else:
                    position-=1
                    if self.board[position]!='-':
                        print('Illegal move; Try again!')
                    else:
                        self.board[position] = self.current_player
                        self.display_board()
                        valid = True
            return

    def check_if_win(self):
        #check rows
        if (self.board[0]==self.current_player and self.board[1]==self.current_player and self.board[2]==self.current_player):
            self.winner = self.current_player
            self.game_still_going = False
            return
        elif (self.board[3]==self.current_player and self.board[4]==self.current_player and self.board[5]==self.current_player):
            self.winner = self.current_player
            self.game_still_going = False
            return
        elif (self.board[6]==self.current_player and self.board[7]==self.current_player and self.board[8]==self.current_player):
            self.winner = self.current_player
            self.game_still_going = False
            return
        #check columns
        elif (self.board[0]==self.current_player and self.board[3]==self.current_player and self.board[6]==self.current_player):
            self.winner = self.current_player
            self.game_still_going = False
            return
        elif (self.board[1]==self.current_player and self.board[4]==self.current_player and self.board[7]==self.current_player):
            self.winner = self.current_player
            self.game_still_going = False
            return
        elif (self.board[2]==self.current_player and self.board[5]==self.current_player and self.board[8]==self.current_player):
            self.winner = self.current_player
            self.game_still_going = False
            return
        #check diagonals
        elif (self.board[0]==self.current_player and self.board[4]==self.current_player and self.board[8]==self.current_player):
            self.winner = self.current_player
            self.game_still_going = False
            return
        elif (self.board[2]==self.current_player and self.board[4]==self.current_player and self.board[6]==self.current_player):
            self.winner = self.current_player
            self.game_still_going = False
            return
        else:
            return

    def check_if_tie(self):
        if (self.winner == None and self.turn!=9):
            return
        elif (self.winner == None and self.turn==9):
            self.game_still_going = False
            return
        else:
            return

    def check_if_game_over(self):
        self.check_if_win()
        self.check_if_tie()
        return

    def play_game(self):
        if self.mode == 1:
            
            #display the initial game
            self.display_board()

            while self.game_still_going:
                if self.current_player=='x': player_name = self.player1.player_name
                else: player_name = self.player2.player_name

                print('{player}\'s turn'.format(player = player_name))
                self.handle_turn()

                self.check_if_game_over()

                if self.game_still_going == False:
                    break

                self.turn+=1
                self.flip_player()

            if (self.winner == 'x' or self.winner == 'o'):
                print(self.winner + ' won')
            elif self.winner == None:
                print('Tie')
        else:
            print('Mode = 2')

            #display the initial game
            self.display_board()
            
            while self.game_still_going:
                if self.current_player=='x': player_name = self.player1.player_name
                else: player_name = self.player2.player_name
                    
                print('{player}\'s turn'.format(player = player_name))
                self.handle_turn()

                self.check_if_game_over()

                if self.game_still_going == False:
                    break

                self.turn+=1

            if (self.winner == 'x' or self.winner == 'o'):
                print(self.winner + ' won')
            elif self.winner == None:
                print('Tie')

                
    class player:
        def __init__(self, name, character):
            self.tokens = 1000
            self.player_name = name
            self.player_character = character #x or o
            self.bet = 0
```
