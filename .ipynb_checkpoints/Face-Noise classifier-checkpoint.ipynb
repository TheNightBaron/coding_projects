```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from tensorflow.keras import Input
from tensorflow.keras.models import (Model, Sequential)
from tensorflow.keras.layers import (Dropout, Dense, Softmax, ReLU, Conv2D)
import tensorflow.keras.layers as layers
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
from sklearn.utils import class_weight
from tensorflow.keras import backend as K
import tensorflow_addons as tfa
import random
```

```python
batch_size = 32
img_height = 224
img_width = 224
```

```python
data_augmentation = tf.keras.Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal", 
                                                 input_shape=(img_height, 
                                                              img_width,
                                                              3)),
    layers.experimental.preprocessing.RandomRotation(0.1),
    layers.experimental.preprocessing.RandomZoom(0.1),
  ]
)
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    path, 
#     labels = 'inferred',
#     label_mode = 'binary',
#     class_names = ['Human','Noise'],
    batch_size = batch_size,
    image_size = (img_height,img_width),
    shuffle = True,
    validation_split=0.2,
    subset="training",
    seed = 69
  )
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    path, #path to folder with subfolders of humans and noises
#     labels = 'inferred',
#     label_mode = 'binary',
#     class_names = ['Human','Noise'],
    batch_size = batch_size,
    image_size = (img_height,img_width),
    shuffle = True,
    validation_split=0.2,
    subset="validation",
    seed = 69
  )
```

```python
for x in train_ds.__iter__():
    print(type(x))
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
labels =  np.array([])
for x, y in train_ds:
    labels = np.concatenate([labels, y])
len(labels), sum(labels)

#classes: 1 is Noise, 0 is Human
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
normalization_layer = layers.experimental.preprocessing.Rescaling(1./255)
```

```python
plt.figure(figsize=(10, 10))
for images, _ in train_ds.take(1):
    for i in range(9):
        augmented_images = data_augmentation(images)
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(augmented_images[0].numpy().astype("uint8"))
        plt.axis("off")
```

```python
num_classes = 2

model = Sequential([
#                     data_augmentation,
                    layers.experimental.preprocessing.Rescaling(1./255, input_shape=(img_height, img_width, 3)),
                    layers.Conv2D(64, 1, activation='relu'),
#                     layers.MaxPooling2D(),
                    layers.Conv2D(64, 3, activation='relu'),
#                     layers.MaxPooling2D(),
#                     layers.Conv2D(64, 5, activation='relu'),
#                     layers.MaxPooling2D(),
                    layers.Flatten(),
                    layers.Dense(16, activation='relu'),
#                     layers.Dropout(0.2),
                    layers.Dense(num_classes, activation = 'softmax')
                   ])
                    
```

```python
model.compile(optimizer='rmsprop',
#               loss=tf.keras.losses.MeanSquaredError(),
              loss = tfa.losses.SigmoidFocalCrossEntropy(from_logits = True, reduction=tf.keras.losses.Reduction.SUM_OVER_BATCH_SIZE),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
class_weights = class_weight.compute_class_weight('balanced',
                                                 np.unique(labels),
                                                 np.array(labels))
```

```python
class_weights = class_weights/sum(class_weights)
```

```python
epochs=3
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs,
#     class_weight = {0:86,1:322}
#     class_weight = {0:class_weights[0],1:class_weights[1]}
)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
img = tf.keras.preprocessing.image.load_img(
    humans_path + '\\' + os.listdir(humans_path)[69], target_size=(img_height, img_width)
)
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch

predictions = model.predict(img_array)
score = tf.nn.softmax(predictions[0])

print(
    "This image most likely belongs to {} with a {:.2f} percent confidence."
    .format(class_names[np.argmax(score)], 100 * np.max(score))
)
```
